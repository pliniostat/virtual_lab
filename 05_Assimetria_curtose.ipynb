## 2) Medidas de assimetria e de curtose

Sejam $x_1,x_2,\ldots,x_n$ um conjunto de dados sobre uma variável quantitativa $X$. Defina:

$$m_2=\frac{1}{n}\sum_{i=1}^n (x_i-\bar{x})^2, \ \ \ \mbox{(variância dos dados)}$$

em que $\bar{x}=\frac{1}{n}\sum_{i=1}^n x_i$ é a média amostral. Defina também a seguinte **padronização** dos dados

$$z_i=\frac{x_i-\bar{x}}{\sqrt{m_2}} \ \ \ (\mbox{para }i=1,2,\ldots,n).$$

Observe que tal padronização faz com que o novo conjunto de dados $z_1,z_2,\ldots,z_n$ tenha média amostral $0$ e variância amostral $1$.

### 2.1) Coeficiente de assimetria de Pearson
O **coeficiente de assimetria de Pearson**, que denotaremos por $b_1$, é definido por:

$$b_1=\frac{1}{n}\sum_{i=1}^n z_i^3.$$

Devemos interpretar o coeficiente de assimetria da seguinte maneira:

* Se $b_1>0$, a distribuição dos dados é assimétrica à **direita**;
* se $b_1<0$, a distribuição dos dados é assimétrica à **esquerda**;
* se $b_1\approx 0$, a distribuição dos dados é simétrica (aproximadamente).

**Observação:** Como $b_1$ depende da média amostral, temos que este coeficiente não é robusto, ou seja, é fortemente influenciado por valores extremos. Note também que tal medida é **adimensional**.

Tal função **não** está implementada no R, mas podemos calcular este coeficiente utilizando as funções que conhecemos (tomando cuidado com os valores faltantes quando estes existirem!).

### 2.2) Coeficiente de assimetria de Bowley

Vimos que as distâncias entre os elementos de um box plot (mediana e quartis) são informativos com respeito à simetria ou assimetria dos dados observados. O **coeficiente de Bowley** explora essas distâncias fornecendo uma medida de assimetria robusta e adimensional. Sua fórmula é dada por:

$$c_B=\frac{(q_3-md)-(md-q_1)}{q_3-q_1},$$

em que $q_1$ e $q_3$ são o primeiro e terceiro quartis e $md$ a mediana dos dados. A interpretação deste coeficiente é a mesma do coeficiente de assimetria de Pearson.

Novamente, podemos calcular este coeficiente facilmente utilizando as funções que já conhecemos.

### 2.3) Coeficiente de curtose
O **coeficiente de curtose** mede o grau de achatamento da distribuição dos dados ou o decaimento das caudas dessa distribuição (peso da cauda). Se o decaimento é lento, dizemos que a distribuição tem caudas pesadas e, neste caso, valores muitos distantes da massa de dados tendem a ocorrer com frequência significativa. Este fenômeno não ocorre quando o decaimento das caudas é rápido, ou seja, quando a distribução dos dados possui caudas leves. 


Dadas as observações padronizadas $z_1,z_2,\ldots,z_n$ definidas no início da seção, o coeficiente curtose (também devido à Karl Pearson) é dado por:

$$b_2=\frac{1}{n}\sum_{i=1}^n z_i^4.$$


Em muitas aplicações, temos que aproximadamente $99,7\%$ das observações estão no intervalo $[\bar{x}-3dp(x), \bar{x}+3dp(x)]$, onde $\bar{x}$ e $dp(x)$ são a média e desvio padrão amostrais respectivamente. Nestes casos, a densidade de frequência fora desse intervalo é de $0,3\%$ (apenas $3$ observações em uma amostra de tamanho $1000$). 



Em situações como essa, o coeficiente de curtose terá valor aproximadamente igual a $3$. Por outro lado, quando a densidade de frequência fora do intervalo $[\bar{x}-3dp(x), \bar{x}+3dp(x)]$ é alta, as caudas da distribuição dos dados é pesada e o coeficiente de curtose será maior do que $3$. Por fim, quando a densidade de frequência fora do intervalo é nula, o coeficiente de curtose será menor do que $3$ e a distribuição dos dados terá caudas leves.

Logo, a curtose da distribuição dos dados pode ser classificada em três tipos quando ao valor de $b_2$:

* Se $b_2<3$, dizemos que a distribuição dos dados é **Platicúrtica** (distribuições achatadas);
* Se $b_2\approx 0$, dizemos que a distribuição dos dados é **Mesocúrtica** (distribuição normal ou gaussiana); 
* Se $b_2>3$, dizemos que a distribuição dos dados é **Leptocúrtica** (distribuições com pico acentuado).

## 2) Pressupostos

Considere duas populações $A$ e $B$ distintas com médias populacionais $\mu_A$ e $\mu_B$ respectivamente. Queremos obter um intervalo de confiança para a diferença $\mu_A-\mu_B$ com coeficiente de confiança $\gamma$.

Para isso, devemos obter uma a.a. $A_1, A_2, \ldots, A_{n_A}$ para a população $A$ e uma a.a. $B_1, B_2, \ldots, B_{n_B}$ para a população $B$.

As técnicas que serão apresentadas a seguir assumem que as amostras aleatórias de cada população **devem ser independentes entre si**.

Para todos os casos a seguir, um I.C. com coeficiente de confiança $\gamma$ para a diferença entre as médias populacionais será dado por:

$$IC(\mu_A-\mu_B,\gamma)=(\bar{D}-E, \bar{D}+E), \ \ \ \ \ \ \mbox{(I)}$$

onde $\bar{D}=\bar{X}_A-\bar{X}_B$ (diferença entre as médias de cada amostra) e $E$ é determinado de acordo com o conhecimento sobre a variabilidade de cada população.

**Em todos os casos, iremos assumir que as populações são normais ou que as amostras são suficientemente grandes.** Essas técnicas ainda funcionam razoavelmente bem para amostras pequenas não-normais, desde que não existam *outliers* ou forte assimetria dos dados. 

## 3) Intervalo de confiança para a diferença entre médias populacionais com variâncias conhecidas

Neste caso, devemos assumir que as variâncias populacionais $\sigma_A^2$ e $\sigma_B^2$ sejam conhecidas. Deste modo, o I.C. dado em $\mbox{(I)}$ é obtido com

$$E=z(\gamma)\sqrt{\frac{\sigma_A^2}{n_A}+\frac{\sigma_B^2}{n_B}},$$

onde $z(\gamma)$ é o quantil de ordem $(1+\gamma)/2$ da distribuição normal padrão.

## 4) Intervalo de confiança para a diferença entre médias populacionais com variâncias desconhecidas porém iguais.

Assuma agora que $\sigma_A^2=\sigma_B^2=\sigma^2$. Sejam $S_A^2$ e $S_B^2$ os estimadores para a variância populacional das populações $A$ e $B$ respectivamente. Mostra-se que o estimador **variância combinada** de $A$ e $B$ dado por

$$S_p^2=\frac{(n_A-1)S_A^2+(n_B-1)S_B^2}{n_A+n_B-2},$$

é não viesado com respeito a $\sigma^2.$

### Exercício 3
Obtenha a variância combinada para os dados de resistência de materiais armazenando-o em um vetor chamado ```sp2```.

Neste caso, o I.C. definido em $\mbox{(I)}$ é obtido com

$$E=t(\gamma)s_p\sqrt{\frac{1}{n_A}+\frac{1}{n_B}},$$

em que $t(\gamma)$ é o quantil de ordem $(1+\gamma)/2$ da distribuição t-Student com $n_A+n_B-2$ graus de liberdade.

Neste caso mais realista, temos que o I.C. definido em $\mbox{(I)}$ é obtido com

$$E=t(\gamma)\sqrt{\frac{s_A^2}{n_A}+\frac{s_B^2}{n_B}},$$

em que $t(\gamma)$ é o quantil de ordem $(1+\gamma)/2$ da distribuição t-Student com $\nu$ graus de liberdade, onde

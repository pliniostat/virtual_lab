## 1) Introduo

Assim como no caso discreto, podemos calcular probabilidades, quantis e simular dados de uma v.a. contnua. Seja ```nome``` o rtulo que o R utiliza para identificar uma distribuio e ```parametros``` os parmetros dessa distribuio. Por exemplo, para uma v.a. binomial, ```nome``` era igual a ```binom``` e ```parametros``` eram os valores ```n,p```. Utiliza-se a mesma sintaxe bsica:

* ```dnome(x, parametros)```: retorna valores da f.d.p. de uma v.a. $X$ no ponto $x$;
* ```pnome(x, parametros)```: retorna valores da f.d.a. de $X$ no ponto $x$;
* ```qnome(alpha, parametros)```: retorna o quantil de ordem $0\leq\alpha\leq 1$ da distribuio de $X$. Ou seja, o valor de $x$ tal que $F(x)=\alpha$;
* ```rnome(N,parametros)```: gera uma amostra de tamanho ```N``` da v.a. $X$.

## 2) Distribuio Qui-quadrado

A distribuio **qui-quadrado** (chi-square em ingls)  uma distribuio muito importante que surge na etapa inferencial de algumas anlises estatsticas como veremos posteriormente.

Sejam $X_1,X_2,\ldots,X_k$ v.a.'s independentes, todas com distribuio $N(0,1)$. Uma v.a. $Q$ com distribuio qui-quadrado surge da seguinte relao:

$$Q=X_1^2+X_2^2+\cdots+X_k^2.$$

Neste caso, dizemos que $Q$ possui distribuio qui-quadrado com $k$ **graus de liberdade**. **Notao:** $Q\sim\chi_k^2.$

Sua f.d.p.  dada por

$$f(x)=\frac{1}{2^{k/2}\Gamma(k/2)}x^{\frac{k}{2}-1}e^{-\frac{x}{2}}, \ \ \ k>0, x>0.$$

O smbolo $\Gamma(\cdot)$ denota a **funo gama**. A figura a seguir ilustra o grfico dessa f.d.p. variando o valor de $k$.



Alm disso, se $Q\sim\chi_k^2$, mostra-se que $\mathbb{E}(Q)=k$ e $\mbox{Var}(Q)=2k$.

Para uma distribuio qui-quadrado com $k$ graus de liberdade, temos as seguintes funes implementadas no R:

* ```dchisq(x, k)```: retorna valores da f.d.p. no ponto $x$;
* ```pchisq(x, k)```: retorna valores da f.d.a. no ponto $x$;
* ```qchisq(alpha, k)```: retorna o quantil de ordem $0\leq\alpha\leq 1$ da distribuio. Ou seja, o valor de $x$ tal que $F(x)=\alpha$;
* ```rchisq(N, k)```: gera uma amostra de tamanho ```N``` com distribuio qui-quadrado com $k$ graus de liberdade.

## 3) Distribuio t-Student

Sejam $Z\sim N(0,1)$ e $Q\sim\chi_k^2$, mostra-se que a v.a.

$$T=\frac{Z}{\sqrt{\frac{Q}{k}}},$$

 tambm uma varivel aleatria cuja distribuio recebe o nome de t-Sudent com **$k$ graus de liberdade**. **Notao:** $T\sim t_k$.

A distribuio **t-Student**, assim como a distribuio anterior, surge frequentemente em problemas de inferncia estatstica. Ela possui caudas mais pesadas que as da distribuio normal, o que faz com que essa distribuio seja mais adequada para lidar com valores discrepantes.

Sua f.d.p.  dada por

$$f(x)=\frac{\Gamma(\frac{k+1}{2})}{\sqrt{k\pi}\Gamma(k/2)}\left(1+\frac{x^2}{k}\right)^{-\frac{k+1}{2}}, \ \ \ k>0,x\in\mathbb{R}.$$

Novamente, $\Gamma(\cdot)$ denota a funo gama. A figura a seguir ilustra o grfico dessa f.d.p. variando o valor de $k$.



A esperana e a varincia de uma v.a. t-Student s est definida para certos valores de $k$. Logo, se $T\sim t_k$, temos que

$$\mathbb{E}(T)=0, \ \ \ \mbox{se} \ \ \ k>1,$$

$$\mbox{Var}(T)=\frac{k}{k-1}, \ \ \ \mbox{se} \ \ \ k>2.$$

Se $k$ for suficientemente grande ($k\rightarrow\infty$), temos que $T$ se aproxima da distribuio $N(0,1)$.

Para uma distribuio t-Student com $k$ graus de liberdade, temos as seguintes funes implementadas no R:

* ```dt(x, k)```: retorna valores da f.d.p. no ponto $x$;
* ```pt(x, k)```: retorna valores da f.d.a. no ponto $x$;
* ```qt(alpha, k)```: retorna o quantil de ordem $0\leq\alpha\leq 1$ da distribuio. Ou seja, o valor de $x$ tal que $F(x)=\alpha$;
* ```rt(N, k)```: gera uma amostra de tamanho ```N``` com distribuio qui-quadrado com $k$ graus de liberdade.
